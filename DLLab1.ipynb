```python
import numpy as np

# Set seed for reproducibility
np.random.seed(0)

# 1.1 Create random 4x4 array
exp_array = np.random.exponential(scale=1.0, size=(4, 4))
print("4x4 Array from Exponential Distribution:\n", exp_array)
```

```python
import matplotlib.pyplot as plt

# Generate 100000 values from each distribution
exp_data = np.random.exponential(scale=1.0, size=100000)
uniform_data = np.random.rand(100000)
normal_data = np.random.randn(100000) #(bell curve shape)

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(exp_data, density=True, bins=100, histtype="step", color="green", label="Exponential")
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(normal_data, density=True, bins=100, histtype="step", color="red", label="Normal")

plt.axis([-2.5, 8, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Create meshgrid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

# Plot 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title("3D Plot of Z = X² + Y²")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Use a built-in dataset (e.g., penguins)
df = sns.load_dataset("penguins")

# Select numeric features
features = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
data = df[features].dropna()

# Pearson and Spearman
pearson_corr = data.corr(method='pearson')
spearman_corr = data.corr(method='spearman')

# Plot heatmaps
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap="Blues")
plt.title("Pearson Correlation")

plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap="Oranges")
plt.title("Spearman Correlation")

plt.tight_layout()
plt.show()
```

```python
#task 2.1
import pandas as pd

file_path = "/content/drive/MyDrive/DL Lab1/Au_nanoparticle_dataset.csv"
df = pd.read_csv(file_path)


# Remove the unwanted columns
new_df = df.drop(['N_total', 'N_bulk', 'N_surface', 'R_avg'], axis=1)

# Show first 5 rows
new_df.head()

```

```python
new_df.head(20)
```

```python
# Mean
mean_values = new_df.mean()
print("Mean values:\n", mean_values)

# Standard deviation
std_values = new_df.std()
print("\nStandard deviations:\n", std_values)

# Quartiles (25%, 50%, 75%)
quartiles = new_df.quantile([0.25, 0.5, 0.75])
print("\nQuartiles:\n", quartiles)
```

```python
#Task 2.6
g = sns.PairGrid(new_df)

# Diagonal: hist + kde
g.map_diag(sns.histplot, kde=True)

# Upper half: bivariate histogram
g.map_upper(sns.histplot)

# Lower half: bivariate KDE
g.map_lower(sns.kdeplot)

plt.suptitle('Customized PairGrid with Histograms and KDEs', y=1.02)
plt.show()
```
